2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of 
release) and make data frame.

```python
from bs4 import BeautifulSoup
import requests
imdb_url = requests.get('https://www.imdb.com/chart/top')
imdb_url
```

```python
soup = BeautifulSoup(imdb_url.content)
soup
```

```python
#Scraped the movie names
scraped_movies=soup.find_all('td',class_="titleColumn")
scraped_movies
```

```python
#parse movie names
movies = [] 
for i in scraped_movies:
    i = i.get_text().replace('\n',"")
    i = i.strip(" ")
    movies.append(i)   
movies    
```

```python
#Scrap rating for movies
scraped_ratings = soup.find_all('td',class_="ratingColumn imdbRating")
scraped_ratings
```

```python
#parse ratings
ratings = [] 
for j in scraped_ratings:
    j = j.get_text().replace('\n',"")
    ratings.append(j)   
ratings    
```

```python
#Store data by using dataframe
import pandas as pd
data = pd.DataFrame()
data['Movie Names & Year'] = movies
data['IMDB Ratings'] = ratings
data.head(100)
```

8. Write a python program to scrape mentioned details from ‘https://www.dineout.co.in/delhi-restaurants/buffet-special’ :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image url

```python
from bs4 import BeautifulSoup
import requests
url = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
url
```

```python
soup = BeautifulSoup(url.content)
soup
```

```python
#scraping data
rest_name=[]
for i in soup.find_all('div',class_='restnt-info cursor'):
    i=i.get_text().replace('\n','')
    i=i.strip(" ")
    i=i.split(',')[0]
    rest_name.append(i)
    
rest_name
```

```python
location=[]
for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    i=i.get_text().replace('\n','')
    i=i.strip(" ")
    i=i.split(',')[1]
    location.append(i)
    
location
```

```python
price=[]
for i in soup.find_all('span',class_='double-line-ellipsis'):
    i=i.get_text().replace('\n','')
    i=i.strip(" ")
    i=i.split('|')[0]
    price.append(i)
    
price
```

```python
cuisine=[]
for i in soup.find_all('span',class_='double-line-ellipsis'):
    i=i.get_text().replace('\n','')
    i=i.strip(" ")
    i=i.split('|')[1]
    cuisine.append(i)
    
cuisine
```

```python
images = []
for i in soup.find_all('img',class_='no-img'):
    images.append(i['data-src'])
    
images
```

```python
import pandas as pd
df = pd.DataFrame()
df['Resturant_name']=rest_name
df['Location']=location
df['Price']=price
df['Cuisine']=cuisine
df['Images_url']=images
df
```

9. Write a python program to scrape weather details for last 24 hours from
‘https://en.tutiempo.net/delhi.html?data=last-24- hours’ :
i) Hour
ii) Temperature
iii) Wind
iv) Weather condition
v) Humidity
vi) Pressure

```python
from bs4 import BeautifulSoup
import requests
url = requests.get('https://en.tutiempo.net/delhi.html?data=last-24-hours')
url
```

```python
soup = BeautifulSoup(url.content)
soup
```

```python
hour = []
for i in soup.find_all('td',class_='t Temp')[0:24]:
    if i.previous_sibling.previous_sibling is not None:
        hour.append(i.previous_sibling.previous_sibling.text)
    else:
        hour.append(' ')

hour
```

```python
#scraping weather condition data
condition_data = soup.find_all('span',class_='thhip ico i0530 u303')
condition_data
```

```python
weather_condition=[]
for i in condition_data:
    i=i.get_text().replace('\n','')
    i=i.strip(" ")
    weather_condition.append(i)
    
weather_condition
```

```python
#scraping temperature data
temp_data = soup.find_all('td',class_='t Temp')
temp_data
```

```python
temperature=[]
for i in temp_data:
    i=i.get_text().replace('\n','')
    i=i.strip(" ")
    temperature.append(i)
    
temperature
```

```python
#scraping wind data
wind_data = soup.find_all('td',class_='wind')
wind_data
```

```python
wind = []
for i in wind_data:
    i=i.get_text().replace('\n','')
    i=i.strip(" ")
    wind.append(i)
    
wind
```

```python
#scraping humidity data
humidity_data = soup.find_all('td',class_='hr')
humidity_data
```

```python
humidity = []
for i in humidity_data:
    i=i.get_text().replace('\n','')
    i=i.strip(" ")
    humidity.append(i)
    
humidity
```

```python
#scraping humidity data
pressure_data = soup.find_all('td',class_='prob')
pressure_data
```

```python
pressure = []
for i in pressure_data:
    i=i.get_text().replace('\n','')
    i=i.strip(" ")
    pressure.append(i)
    
pressure
```

```python
import pandas as pd
df = pd.DataFrame()
df['Hour']=hour
df['Weather_condition']=weather_condition
df['Temperature']=temperature
df['Wind']=wind
df['Humidity']=humidity
df['Pressure']=pressure
df
```

```python
soup = BeautifulSoup(url.content)
soup
```

```python
#scraping data
data = soup.find_all('div',class_='blog--single__content column--3-4 u-spacing-third')
data
```

```python
temp=[]
for i in data:
    for j in i.find_all('p'):
        t= j.text.replace('\n',"")
        temp.append(t)
```

```python
name = []
description = []
for i in range(1,30,3):
    name.append(temp[i])
for i in range(2,30,3):
    description.append(temp[i])
```

```python
name
```

```python
description
```

```python
#scraping image
temp = []
for i in soup.find_all('img'):
    temp.append(i['src'])
```

```python
img_url = []
for i in range(6,26,2):
    img_url.append(temp[i])

img_url
```

```python
import pandas as pd
df = pd.DataFrame()
df['Monument_Name']=name
df['Description']=description
df['Image_url']=img_url
df
```
